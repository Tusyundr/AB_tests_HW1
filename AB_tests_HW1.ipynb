## Задание 1. Выберите любой продукт.
Например ваш продукт

* ozon.ru

* citilink.ru

* telegram

Какие бы ключевые метрики вы выделили? Опишите почему вы выбрали именно их и можно ли их измерить.

Выбранный продукт: **citilink.ru**

Ключевые метрики:
1. **Количество посетителей веб-сайта**. Показывает насколько интернет-магазин виден в сети, а ассортимент товара востребован. Данную метрику можно измерить с помощью инструментов веб-аналитики.

2. **Количество страниц, просматриваемых пользователем на сайте.** Гипотетически, если посетитель заинтересован в продукции и готов к покупке, он будет просматривать большое количество страниц. Данную метрику можно измерить с помощью инструментов веб-аналитики.

3. **Доля вернувшихся пользователей.** Показывает уровень удержания клиентов. Данную метрику можно измерить с помощью инструментов веб-аналитики.


## Задание 2. Сформулируйте 5-10 гипотез для выбранного продукта.

1.  Если сделать доставку дешевле или бесплатной, это может увеличить количество продаж, а также долю вернувшихся пользователей, потому что доставка станет доступнее, а значит покупка выгоднее.

2. Если сделать более разнообразными фильтры поиска и где-то боку или внизу странцы отбражать товары, которые на основании  этих фильтров могут быть интересны покпателю, это может повысить число продаж, потому что клиентам будет легче найти именно то, что они хотят купить.

3.  Оптимизация сайта интернет-магазина под мобильные устройства может повысить среднее время пребывания на сайте, глубину просмотра страниц и  число продаж, так как большая доля людей использует смартфоны для поиска и покупки товаров.

4. Если добавить возможность совершать покупку без регистрации, то может повысится число продаж, поскольку клиенты будут тратить меньше времени на процесс покупки.

5. Расширение ассортимента товаров может увеличить среднее время пребывания на сайте и количество просматриваемых страниц, так как расширение ассортимента приведет к увеличению количества потенциальных покупателей.

6. Увеличение  кешбэка бонусами моржет привести к росту среднего чека, поскольку фактически это увеличит размер скидки на продукты (оплата накопленными бонусами), что позволит клиентам покупть более дорогие позиции.
